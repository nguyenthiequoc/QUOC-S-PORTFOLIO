**Hình 1: Có mối tương quan như thế nào giữa trọng lượng và mức tiêu thụ nhiên liệu (mpg) của các mẫu xe?**

**Hình 7: Dữ liệu cho thấy có mối tương quan nghịch giữa gia tốc và MPG, tức là khi gia tốc cao thì MPG thường thấp đúng hay sai? Hãy dựa vào các phân bố và mối quan hệ giữa các biến để giải thích.**

```python
import pandas as pd
import plotly.figure_factory as ff
import numpy as np
x1 = pd.to_numeric(mpg_df['acceleration'], errors='coerce').dropna()
x2 = pd.to_numeric(mpg_df['displacement'], errors='coerce').dropna()
x3 = pd.to_numeric(mpg_df['mpg'], errors='coerce').dropna()
hist_data = [x1, x2, x3]
group_labels = ['Acceleration', 'Displacement', 'MPG']
colors = ['#0165fc', '#0cff0c', '#00ffff']
fig = ff.create_distplot(hist_data, group_labels, colors=colors, bin_size=.5, show_curve=True,show_hist=True)
fig.show()
```

Nhận xét:
* Phân bố của biến gia tốc trong biểu đồ thể hiện rõ điều này. Có thể quan sát thấy một số điểm gia tốc vượt trội so với phần còn lại, lên tới khoảng 0.15 - 0.2g. Những giá trị gia tốc cực đại này có thể là do những tình huống lái xe đặc biệt hoặc do đặc tính vận hành của phương tiện.
* Khi xem xét phân bố của biến MPG, ta thấy các giá trị MPG tập trung chủ yếu trong khoảng 0.02 - 0.08, với một số điểm ở giá trị cao hơn. Nhìn chung, phân bố MPG có xu hướng giảm khi gia tốc tăng cao.
* Quan sát trực quan mối quan hệ giữa gia tốc và MPG trong biểu đồ cũng cho thấy rõ ràng có sự tương quan nghịch giữa hai biến số này. Khi gia tốc cao, thường tương ứng với những giá trị MPG thấp hơn. Điều này hoàn toàn phù hợp với nhận định ban đầu.
* Dựa trên các đặc điểm phân bố của gia tốc và MPG, cũng như mối quan hệ giữa chúng thể hiện trong biểu đồ, ta có thể khẳng định rằng nhận định "Dữ liệu cho thấy có mối tương quan nghịch giữa gia tốc và MPG, tức là khi gia tốc cao thì MPG thường thấp" là hoàn toàn chính xác.

Một số mẫu xe có đường cong đạt giá trị đỉnh cao hơn so với những mẫu khác, cho thấy chúng có hiệu suất nhiên liệu tốt hơn. Sự thay đổi hình dạng và vị trí của các đường cong cũng phản ánh diễn biến của hiệu suất nhiên liệu theo từng năm.
